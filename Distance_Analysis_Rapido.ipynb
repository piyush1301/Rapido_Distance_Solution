```python
# Importing the modules 
import numpy as np
import pandas as pd
#Reading the file
sample=pd.read_csv('Position.csv')
#Filtering the Data
sample=sample[['user_id','coordinate']]
#Grouping the Data to get distance for each user_id
smap_grp=sample.groupby('user_id')
#Creating a Flag to control my output and reset my distance for each user_id
dir_={}
for i in sample['user_id'].unique():
    dir_[i]=0
#Creating the answer DataFrame to save the answer for each user_id
Ans=pd.DataFrame([])
#Function to calculate the result for each user_id
def count_dis(row):
    #taking global variable where countx is the last x value and county is last y value
    global countx;    global county;    global total
    #featch the user id
    i=row['user_id']
    #if this is first occuring of the user_id then --
    if dir_[i]==0:
        # reset the value of dir_ and reset the distance to 0,0 
        dir_[i]=1;        countx=0;        county=0;        total=0
    #featch the coordinate for row
    row['coordinate']
    # split the x,y variable and then map them to integer type
    x,y=list(map(int, row['coordinate'].split(',')))
    # applying the formula 
    total=total+abs(x-countx)+abs(y-county)
    # creatring the last value to save into global variable
    countx,county=x,y  
    # saving the answer
    Ans.loc[i,'distance']=total
#Applying my function to each data set
countx=0
county=0
total=0
for i,j in smap_grp:
    smap_grp.get_group(i).apply(lambda row: count_dis(row),axis=1);
#reset the index and renaming the columns
Ans.reset_index(inplace=True)
Ans.columns=['user_id','distance']
#Saving the result 
Ans.to_csv('Solution.csv',index=False)
```
